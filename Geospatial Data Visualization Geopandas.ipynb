```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
world_map = gpd.read_file(r"C:\Users\jh37495\Geospatial Data Science\Geospatial Data Visualization using GeoPandas\world.shp")
```

```python
world_map
```

```python
world_map = world_map[['NAME', 'geometry']]
world_map
```

```python
world_map['area'] = world_map.area
```

```python
world_map
```

```python
world_map.plot()
```

```python
# Remove Antarctica from GeoPandas GeoDataFrame
world_map = world_map[world_map["NAME"] != 'Antarctica']
world_map.plot(figsize = (10, 8))
```

```python
# Change the projection (WGS84: degree -> meter)
world_map.crs
```

```python
world_map.to_crs(epsg = 3857, inplace = True)
world_map.crs
```

```python
# Recaluating areas in Sq. Km.
world_map['area'] = world_map.area/1000000
```

```python
# Adding legend to world_map
world_map.plot(column = 'area', cmap = 'Greens', legend = True,
              legend_kwds = {'label': 'Area of the country (Sq. Km.)'},
              figsize = (10, 8))
```

```python
# Resize the legend
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots(figsize = (10, 8))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size = "7%", pad = 0.1)

world_map.plot(column = 'area', cmap = 'Greens', legend = True,
              legend_kwds = {'label': 'Area of the country (Sq. Km.)'},
              ax = ax, cax = cax)
```
